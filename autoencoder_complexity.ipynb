```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets
import sklearn.preprocessing
import torch
import torchvision
import torchvision.datasets
import torchvision.models
import torchvision.transforms
```

```python
numeros = sklearn.datasets.load_digits()
imagenes = numeros['images']  # Hay 1797 digitos representados en imagenes 8x8
n_imagenes = len(imagenes)
X = imagenes.copy()
Y = numeros['target']
print(np.shape(X))
inputs = torch.autograd.Variable(torch.Tensor(X).float()).unsqueeze(1)
print(inputs.size())

n_items = inputs.data.size()[0]
random_items = np.random.choice(np.arange(n_items), 5)

plt.figure(figsize=(14,5))
for i in range(5):
    plt.subplot(1,5,i+1)
    item  = random_items[i]
    plt.imshow(inputs[item][0].detach().numpy())
    plt.title(Y[item])

mean = inputs.mean(dim=0)
std = inputs.std(dim=0)
std[std==0]=1.0

for i in range(len(inputs)):
    inputs[i] = (inputs[i])/std
np.shape(inputs)
```

```python
class A1(torch.nn.Module):
    def __init__(self):
        super(A1,self).__init__()
        self.encoder = torch.nn.Sequential(
            torch.nn.Conv2d(1, 4, kernel_size=3),
            torch.nn.Conv2d(4,8,kernel_size=3),
            torch.nn.Conv2d(8,4,kernel_size=3))
        self.decoder = torch.nn.Sequential(
            torch.nn.ConvTranspose2d(4,8,kernel_size=3),
            torch.nn.ConvTranspose2d(8,4,kernel_size=3),
            torch.nn.ConvTranspose2d(4,1,kernel_size=3))
    def forward(self,x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x
    
class A2(torch.nn.Module):
    def __init__(self):
        super(A2,self).__init__()
        self.encoder = torch.nn.Sequential(
            torch.nn.Conv2d(1, 4, kernel_size=2),
            torch.nn.Conv2d(4,8,kernel_size=2),
            torch.nn.Conv2d(8,4,kernel_size=2))
        self.decoder = torch.nn.Sequential(
            torch.nn.ConvTranspose2d(4,8,kernel_size=2),
            torch.nn.ConvTranspose2d(8,4,kernel_size=2),
            torch.nn.ConvTranspose2d(4,1,kernel_size=2))
    def forward(self,x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x

class A3(torch.nn.Module):
    def __init__(self):
        super(A3,self).__init__()
        self.encoder = torch.nn.Sequential(
            torch.nn.Conv2d(1, 4, kernel_size=1),
            torch.nn.Conv2d(4,8,kernel_size=1),
            torch.nn.Conv2d(8,4,kernel_size=1))
        self.decoder = torch.nn.Sequential(
            torch.nn.ConvTranspose2d(4,8,kernel_size=1),
            torch.nn.ConvTranspose2d(8,4,kernel_size=1),
            torch.nn.ConvTranspose2d(4,1,kernel_size=1))
    def forward(self,x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x

class A4(torch.nn.Module):
    def __init__(self):
        super(A4,self).__init__()
        self.encoder = torch.nn.Sequential(
            torch.nn.Conv2d(1, 4, kernel_size=2),
            torch.nn.Conv2d(4,8,kernel_size=2))
        self.decoder = torch.nn.Sequential(
            torch.nn.ConvTranspose2d(8,4,kernel_size=2),
            torch.nn.ConvTranspose2d(4,1,kernel_size=2))
    def forward(self,x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x

class A5(torch.nn.Module):
    def __init__(self):
        super(A5,self).__init__()
        self.encoder = torch.nn.Sequential(
            torch.nn.Conv2d(1, 15, kernel_size=3),
            torch.nn.Conv2d(15,7,kernel_size=3))
        self.decoder = torch.nn.Sequential(
            torch.nn.ConvTranspose2d(7,15,kernel_size=3),
            torch.nn.ConvTranspose2d(15,1,kernel_size=3))
    def forward(self,x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x
###############
class A6(torch.nn.Module):
    def __init__(self):
        super(A6,self).__init__()
        self.encoder = torch.nn.Sequential(
            torch.nn.Conv2d(1, 4, kernel_size=2),
            torch.nn.Conv2d(4,8,kernel_size=2),
            torch.nn.Conv2d(8,4,kernel_size=2))
        self.decoder = torch.nn.Sequential(
            torch.nn.ConvTranspose2d(4,8,kernel_size=2),
            torch.nn.ConvTranspose2d(8,4,kernel_size=2),
            torch.nn.ConvTranspose2d(4,1,kernel_size=2))
    def forward(self,x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x
    
class A7(torch.nn.Module):
    def __init__(self):
        super(A7,self).__init__()
        self.encoder = torch.nn.Sequential(
            torch.nn.Conv2d(1, 4, kernel_size=2),
            torch.nn.Conv2d(4,8,kernel_size=2),
            torch.nn.Conv2d(8,16,kernel_size=2),
            torch.nn.Conv2d(16,8,kernel_size=2))
        self.decoder = torch.nn.Sequential(
            torch.nn.ConvTranspose2d(8,16,kernel_size=2),
            torch.nn.ConvTranspose2d(16,8,kernel_size=2),
            torch.nn.ConvTranspose2d(8,4,kernel_size=2),
            torch.nn.ConvTranspose2d(4,1,kernel_size=2))
    def forward(self,x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x

class A8(torch.nn.Module):
    def __init__(self):
        super(A8,self).__init__()
        self.encoder = torch.nn.Sequential(
            torch.nn.Conv2d(1, 4, kernel_size=1),
            torch.nn.Conv2d(4,1,kernel_size=1))
        self.decoder = torch.nn.Sequential(
            torch.nn.ConvTranspose2d(1,4,kernel_size=1),
            torch.nn.ConvTranspose2d(4,1,kernel_size=1))
    def forward(self,x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x

class A9(torch.nn.Module):
    def __init__(self):
        super(A9,self).__init__()
        self.encoder = torch.nn.Sequential(
            torch.nn.Conv2d(1, 4, kernel_size=3),
            torch.nn.Conv2d(4,8,kernel_size=3))
        self.decoder = torch.nn.Sequential(
            torch.nn.ConvTranspose2d(8,4,kernel_size=3),
            torch.nn.ConvTranspose2d(4,1,kernel_size=3))
    def forward(self,x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x

class A10(torch.nn.Module):
    def __init__(self):
        super(A10,self).__init__()
        self.encoder = torch.nn.Sequential(
            torch.nn.Conv2d(1, 15, kernel_size=1),
            torch.nn.Conv2d(15,7,kernel_size=1))
        self.decoder = torch.nn.Sequential(
            torch.nn.ConvTranspose2d(7,15,kernel_size=1),
            torch.nn.ConvTranspose2d(15,1,kernel_size=1))
    def forward(self,x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x
```

```python
N_C = []
loss_min = []
N_L = []

for M in [A1,A2,A3,A4,A5,A6,A7,A8,A9,A10]:
    num_epochs = 100
    model = M()
    print(M)
    distance = torch.nn.MSELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001, weight_decay=1E-3)

    n_p = 0
    for m in model.parameters():
        n_p += m.flatten().size()[0]
    N_C.append(n_p)
    
    loss_list = [] 
    for epoch in range(num_epochs):
        output = model(inputs)
        loss = distance(output, inputs)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        loss_list.append(loss.item())
        #print('epoch [{}/{}], loss:{:.4f}'.format(epoch+1, num_epochs, loss.item()))
    loss_min.append(min(loss_list))
    
    x_transform = model(inputs)
    latent_space = model.encoder(inputs)
    #print(latent_space.size())
    #print(latent_space.size()[2])
    N_L.append(latent_space.size()[2])
```

```python
plt.figure(figsize=(14,5))

plt.subplot(1,3,1)
plt.scatter(N_C,loss_min)
plt.ylabel('loss')
plt.xlabel(r'$N_C$')

plt.subplot(1,3,2)
plt.scatter(N_L,loss_min)
plt.ylabel('loss')
plt.xlabel(r'$N_L$')

plt.subplot(1,3,3)
plt.scatter(N_C, N_L)
plt.xlabel(r'$N_C$')
plt.ylabel(r'$N_L$')
plt.savefig('graficas.png')
```
